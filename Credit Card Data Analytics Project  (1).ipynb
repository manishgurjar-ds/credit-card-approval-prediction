```python
data=data[["Car_Owner","Propert_Owner","Annual_income","EDUCATION","label"]]
```

```python
data.head()
```

```python
from sklearn import preprocessing
labelencoder=preprocessing.LabelEncoder()
data["Car_Owner"]=labelencoder.fit_transform(data["Car_Owner"])
data["Propert_Owner"]=labelencoder.fit_transform(data["Propert_Owner"])
data["EDUCATION"]=labelencoder.fit_transform(data["EDUCATION"])
```

```python
data.head()
```

```python
data["EDUCATION"].value_counts()
```

```python
X=data.iloc[:,:-1]
```

```python
X.head()
```

```python
Y= data.iloc[:,-1]
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,Y, test_size=0.2)
```

```python
from sklearn.preprocessing import  StandardScaler
#StandardScaler() â†’ numeric features ko 0 mean aur 1 standard deviation me normalize karta hai, taki ML model efficiently aur unbiased train ho.
```

```python
scaler= StandardScaler()
scaler_X_train= scaler.fit_transform(X_train)
scaler_X_test=scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
def modelpeformance(prediction):
    print("Accuracy_Score on model is {}".format(accuracy_score(y_test, prediction)))
                
```

```python
from sklearn.linear_model  import LogisticRegression
Log_model= LogisticRegression()
Log_model.fit(scaler_X_train,y_train)
```

```python
Log_predictions=Log_model.predict(scaler_X_test)
modelpeformance(Log_predictions)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors  import KNeighborsClassifier

param_grid = {
    "n_neighbors": [3, 5, 7, 9, 11],
    "weights": ["uniform", "distance"],   # 'weight' â†’ sahi parameter name 'weights'
    "algorithm": ["auto", "kd_tree"]      # spelling correct
}
gridkm=GridSearchCV(KNeighborsClassifier(),param_grid,cv=5)
```

```python
gridkm.fit(scaler_X_train,y_train)
```

```python
knpreds=gridkm.predict(scaler_X_test)
modelpeformance(knpreds)
```

```python
from sklearn.svm import SVC
svm_model=SVC()
```

```python
param_garid_svr = {"C":[0.01,0.1,0.5,1],
           "kernel": ["linear","rbf","poly"]}
gridsvr = GridSearchCV(svm_model , param_garid_svr)
gridsvr.fit(scaler_X_train, y_train)
```

```python
pred_svc=gridsvr.predict(scaler_X_test)
modelpeformance(pred_svc)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
param_grid = {
    "criterion": ["gini", "entropy"],
    "splitter": ["best", "random"],
    "max_depth": [None, 10, 20, 30, 40, 50],
    "min_samples_split": [2, 5, 10]
}

grid_search_tree = GridSearchCV(estimator=DecisionTreeClassifier(), param_grid=param_grid, cv=5)
grid_search_tree.fit(scaler_X_train, y_train)
```

```python
grid_search_tree.best_params_
```

```python
preds=grid_search_tree.predict(scaler_X_test)
```

```python
modelpeformance(preds)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model=RandomForestClassifier()
```

```python
# Parameter grid
param_grid_rfc = {
    'n_estimators': [32, 64, 128, 256],
    'max_features': [2, 3, 4],        # or ['auto', 'sqrt', 'log2']
    'bootstrap': [True, False],
    'oob_score': [True, False]        # only works when bootstrap=True
}

# GridSearchCV
grid_rfc = GridSearchCV(rfc_model, param_grid_rfc, cv=5)
```

```python
grid_rfc.fit(scaler_X_train,y_train)
```

```python
grid_rfc.best_params_

```

```python
predsrfcc=grid_rfc.predict(scaler_X_test)
modelpeformance(predsrfcc)
```
